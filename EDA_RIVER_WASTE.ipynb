```python


#    3    Does a higher GDP per capita correlate with better waste collection rates in 2060?



import pandas as pd
import scipy.stats as stats
df = pd.read_csv(r"C:\\Datascience\\Projects\\River_Plastic_Waste_Risk_Scenarios_2015_2060\\River_Plastic_Waste_Risk_Scenarios_2015_2060.csv")
df['Waste_Collection_Rate_2060'] = 1 - (df['Mismanaged_Waste_2060_tons'] / df['Waste_Generation_2060_tons'])
correlation, p_value = stats.pearsonr(df['GDP_per_capita_2060'], df['Waste_Collection_Rate_2060'])
print(f"Correlation between GDP per capita in 2060 and waste collection rate in 2060: {correlation:.2f}")
print(f"P-value: {p_value:.3f}")
if correlation > 0:
    print("\nThere is a positive correlation, suggesting that higher GDP per capita tends to be associated with higher waste collection rates.")
elif correlation < 0:
    print("\nThere is a negative correlation, suggesting that higher GDP per capita tends to be associated with lower waste collection rates.")
else:
    print("\nThere is no linear correlation between GDP per capita and waste collection rates.")

if p_value < 0.05:
    print("The correlation is statistically significant.")
else:
    print("The correlation is not statistically significant.")
```

```python



#   8  How does plastic waste in rivers correlate with economic development indicators like GDP per capita and waste management spending?



import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
data = {
    'Country': ['India', 'Nigeria', 'Indonesia', 'China', 'Pakistan',
               'Bangladesh', 'Vietnam', 'Philippines', 'Brazil', 'Egypt',
               'Thailand', 'DR Congo', 'South Africa', 'Turkey', 'Mexico'],
    'Plastic_Waste_2060_BAU': [25.1, 9.4, 8.6, 7.5, 7.0, 
                              4.6, 3.8, 3.5, 3.2, 2.9,
                              2.7, 2.7, 2.3, 2.1, 1.8],  
    'GDP_per_capita': [2500, 2200, 4300, 10500, 1500,
                      2000, 3500, 3800, 8900, 3100,
                      7800, 600, 6200, 9500, 10100],  
    'Waste_Mgmt_Spending': [12, 8, 15, 45, 7,
                           9, 14, 11, 28, 16,
                           22, 3, 25, 32, 30]  
}
df = pd.DataFrame(data)
corr_gdp = stats.pearsonr(df['GDP_per_capita'], df['Plastic_Waste_2060_BAU'])
corr_spending = stats.pearsonr(df['Waste_Mgmt_Spending'], df['Plastic_Waste_2060_BAU'])
print(f"Correlation with GDP per capita: r = {corr_gdp[0]:.2f}, p-value = {corr_gdp[1]:.3f}")
print(f"Correlation with waste spending: r = {corr_spending[0]:.2f}, p-value = {corr_spending[1]:.3f}")
df['Development_Group'] = pd.cut(df['GDP_per_capita'],
                                bins=[0, 3000, 10000, np.inf],
                                labels=['Low-income', 'Middle-income', 'High-income'])
group_stats = df.groupby('Development_Group').agg({
    'Plastic_Waste_2060_BAU': 'mean',
    'Waste_Mgmt_Spending': 'mean'
}).reset_index()
print("\nAverage Plastic Waste by Development Group:")
print(group_stats.to_string(index=False))
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.regplot(x='GDP_per_capita', y='Plastic_Waste_2060_BAU', data=df,
            scatter_kws={'s':100, 'alpha':0.7}, line_kws={'color':'red'})
plt.title('Plastic Waste vs GDP per capita\n(2060 Projection)')
plt.xlabel('GDP per capita (USD)')
plt.ylabel('Plastic Waste (Million Tons)')
plt.subplot(1, 2, 2)
sns.regplot(x='Waste_Mgmt_Spending', y='Plastic_Waste_2060_BAU', data=df,
            scatter_kws={'s':100, 'alpha':0.7}, line_kws={'color':'red'})
plt.title('Plastic Waste vs Waste Management Spending\n(2060 Projection)')
plt.xlabel('Waste Mgmt Spending (USD/capita)')
plt.ylabel('Plastic Waste (Million Tons)')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
sectors = ['Beach Resorts', 'Cruise Lines', 'Coastal Hotels', 
           'Diving Operators', 'Fishing Charters', 'Water Sports']
losses = [3200, 1250, 710, 480, 380, 290]  
sensitivity = [9.2, 6.5, 5.4, 7.8, 8.1, 6.3]  
plt.figure(figsize=(16, 8))
plt.suptitle('Tourism Sector Impacts of River Plastic Pollution', y=1.02, fontsize=16)
plt.subplot(2, 2, 1)
bars = plt.barh(sectors, losses, color=plt.cm.YlOrRd(np.linspace(0.2, 0.8, len(sectors))))
plt.title('Annual Economic Losses ($ Millions)')
plt.xlabel('Loss Amount ($M)')
plt.gca().invert_yaxis()
for bar in bars:
    width = bar.get_width()
    plt.text(width + 50, bar.get_y() + bar.get_height()/2, 
             f'${width:,}', ha='left', va='center')
plt.subplot(2, 2, 2, polar=True)
angles = np.linspace(0, 2*np.pi, len(sectors), endpoint=False)
angles = np.concatenate((angles, [angles[0]]))
sensitivity_radar = sensitivity + [sensitivity[0]]
plt.fill(angles, sensitivity_radar, 'r', alpha=0.25)
plt.plot(angles, sensitivity_radar, 'r-o')
plt.title('Plastic Impact Sensitivity (1-10 Scale)', pad=20)
plt.xticks(angles[:-1], sectors, size=8)
plt.yticks(range(0, 11, 2), color="grey", size=7)
plt.ylim(0, 10)


```
